```python
# Calculate mean, median, and mode of the ratings
mean_rating = data['Rating'].mean()
median_rating = data['Rating'].median()
mode_rating = data['Rating'].mode()[0]  # mode returns a Series, so we take the first element

print(f"Mean Rating: {mean_rating}")
print(f"Median Rating: {median_rating}")
print(f"Mode Rating: {mode_rating}")
```

```python
# Calculate mean, median, and mode of the ratings
mean_rating = data['Rating'].mean()
median_rating = data['Rating'].median()
mode_rating = data['Rating'].mode()[0]  # mode returns a Series, so we take the first element

print(f"Mean Rating: {mean_rating}")
print(f"Median Rating: {median_rating}")
print(f"Mode Rating: {mode_rating}")
```

```python
# Calculate range and IQR for Purchase_Amount
purchase_range = data['Purchase_Amount'].max() - data['Purchase_Amount'].min()
iqr_purchase = data['Purchase_Amount'].quantile(0.75) - data['Purchase_Amount'].quantile(0.25)

print(f"Range of Purchase Amount: {purchase_range}")
print(f"IQR of Purchase Amount: {iqr_purchase}")
```

```python
# Calculate variance and standard deviation for the number of likes
variance_likes = data['Likes'].var()
std_likes = data['Likes'].std()

print(f"Variance of Likes: {variance_likes}")
print(f"Standard Deviation of Likes: {std_likes}")
```

```python
from scipy import stats

# Filter the ratings for Instagram and WhatsApp
instagram_ratings = data[data['App_Name'] == 'Instagram']['Rating']
whatsapp_ratings = data[data['App_Name'] == 'WhatsApp']['Rating']

# Ensure there are no missing values in the ratings
instagram_ratings = instagram_ratings.dropna()
whatsapp_ratings = whatsapp_ratings.dropna()

# Perform a one-tailed independent t-test
t_stat, p_value = stats.ttest_ind(instagram_ratings, whatsapp_ratings, alternative='greater')

# Output the t-statistic and p-value
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Conclusion based on p-value
alpha = 0.05  # Significance level of 5%
if p_value < alpha:
    print("Reject the null hypothesis: Instagram's average rating is significantly higher than WhatsApp's average rating.")
else:
    print("Fail to reject the null hypothesis: No significant difference between Instagram and WhatsApp ratings.")
```

```python
from IPython.display import display, HTML

display(HTML("""
<h2 style='color:#2c3e50;'>7. Sampling Distribution and Central Limit Theorem</h2>

<p style='font-size:15px; line-height:1.7; text-align:justify;'>
<b>Task:</b> Draw multiple random samples of ratings from the dataset and compute the 
<b>mean rating</b> for each sample. Use these sample means to create a 
<b>sampling distribution</b>.
</p>

<p style='font-size:15px; line-height:1.7; text-align:justify;'>
<b>Question:</b> Explain how this sampling distribution demonstrates the 
<b>Central Limit Theorem (CLT)</b>—that the distribution of sample means becomes 
<b>approximately normal</b> as the sample size increases, even when the original data is not normally distributed.
</p>
"""))
```

```python
import numpy as np

# Take random samples of ratings and calculate their means
sample_means = []
for _ in range(1000):  # take 1000 samples
    sample = np.random.choice(data['Rating'], size=30)  # sample of size 30
    sample_means.append(np.mean(sample))

# Plot the sampling distribution
sns.histplot(sample_means, kde=True)
plt.title('Sampling Distribution of the Mean Rating')
plt.xlabel('Mean Rating')
plt.ylabel('Frequency')
plt.show()
```
